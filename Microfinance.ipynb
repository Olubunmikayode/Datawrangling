```python
#confirm if there are outliers
def outliers_no():
  for col in columns:
    Q1 = df_new[col].quantile(0.25)
    Q3 = df_new[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    no_of_outliers = df_new[(df_new[col] < lower_bound) | (df_new[col] > upper_bound)].shape[0]
    return no_of_outliers

columns = ['Age', 'TransactionAmount', 'InterestRate', 'LoanTerm', 'LateFees']

for col in columns:

    print(f'There are {outliers_no()} outliers in {col}')
```

# Data Analysis


# 1.   What are the most common demographic profiles of clients taking loans

```python
df_new.head()
```

```python
df_new['Age'].describe()
```

```python
df_new['Age'].nunique()
```

```python
def age_map(x):
  if x <= 25:
    return '18-25'
  elif x >=26 and x <= 35:
    return '26-35'
  elif x >=36 and x <= 45:
    return '36-45'
  elif x >=46 and x <= 55:
    return '46-55'
  elif x >=56 and x <= 65:
    return '56-65'
  else:
    return '65+'

df_new['Age_Group'] = df_new['Age'].apply(age_map)
```

```python
df_new.head()
```

```python
#using groupby function to create the stacked chart
age_gend = df_new.groupby(['Age_Group', 'Gender']).size().unstack(fill_value=0)
age_gend
```

```python
age_gend['Total'] = age_gend.sum(axis =1)
age_gend
```

```python
age_gend.sort_values(by='Total', ascending=False, inplace =True)
age_gend
```

```python
age_gend= age_gend.drop(columns =['Total'])
age_gend
```

```python
colors = {'Male':'purple', 'Female':'pink'}
ax = age_gend.plot(kind = 'bar', stacked = True, color =[colors[col] for col in age_gend.columns])

for container in ax.containers:
  ax.bar_label(container, label_type = 'center')

plt.title('Demographics of clients by age and gender \n')
plt.xlabel('Age group')
plt.ylabel('Count \n\n', rotation = 45)
plt.xticks(rotation = 45)
plt.legend(title=None)
```

**Most of those taking loans are within the age group of 36 to 45 year and are mostly females**

```python
pivot_table = pd.pivot_table(df_new, index = 'Location', columns = 'Branch', aggfunc = 'size', fill_value = 0)
pivot_table
```

```python
sns.heatmap(pivot_table, cmap = 'Purples', annot= True, fmt = '.1f')
plt.title('Demographics of clients by Location and Branch\n')
plt.ylabel('Location\n\n', rotation = 45)
plt.xlabel('Branch')
plt.yticks(rotation = 45)
```

```python
ag = df_new['Gender'].value_counts()
labels = ag.index
sizes = ag.values
colours = {'Male':'purple', 'Female':'pink'}

plt.figure(figsize=(3, 3))
plt.pie(sizes, labels = labels, autopct = '%1.1f%%', startangle = 120, wedgeprops = {'width':0.3}, colors =[colours[col] for col in ag.index])
plt.title('Gender of clients')
```

```python
ah = df_new['Age_Group'].value_counts()


labels = ah.index
sizes = ah.values
colors = sns.color_palette('pastel')

plt.figure(figsize=(3, 3))
plt.pie(sizes, labels = labels, autopct = '%1.1f%%', colors = colors, startangle = 140)
plt.title('Age group of clients \n')
```

**The branches have about the same patronage**


# 2.   What is the average loan amount given to clients and how does it vary by demographic

```python
df_new['TransactionType'].unique()
```

```python
loan_disbursed = df_new[df_new['TransactionType'] =='Loan Disbursement']
loan_disbursed
```

```python
pivot_table1 = pd.pivot_table(loan_disbursed, index = 'Age_Group', columns='Gender', values = 'TransactionAmount', aggfunc = 'mean')
pivot_table1
```

```python
pivot_table1['Total'] = pivot_table1.sum(axis = 1)
pivot_table1 = pivot_table1.sort_values('Total', ascending=False).drop(columns = ['Total'])
pivot_table1 = pivot_table1.round(2)
pivot_table1
```

```python
colours = {'Female':'pink', 'Male': 'purple'}

plt.figure(figsize=(16, 12))
pivot_table1.plot(kind = 'bar', color = [colours[col] for col in pivot_table1.columns])


plt.title('Average loan amount given to clients by demographic \n')
plt.xlabel('Age group')
plt.ylabel('Average loan amount \n\n\n\n', rotation = 45)
plt.xticks(rotation = 45)
plt.legend(title=None, loc = 'upper left')
```

```python
pivot_table2 = pd.pivot_table(loan_disbursed, index = 'Location', columns='Branch', values = 'TransactionAmount', aggfunc = 'mean')
pivot_table2
```

```python
pivot_table2 = pivot_table2.round(2)
pivot_table2
```

```python
sns.heatmap(pivot_table2, cmap = 'Purples', annot= True, fmt = '.1f')
plt.title('Average loan amount given by Location and Branch \n')
plt.ylabel('Location\n\n', rotation = 45)
plt.xlabel('Branch')
plt.yticks(rotation = 45)
```
